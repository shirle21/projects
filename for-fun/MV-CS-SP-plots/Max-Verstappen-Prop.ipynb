```python
import pandas as pd
```

```python
#Mac Verstappen vs Carlos Sainz 2015 data
mvcspts = pd.read_csv('MaxCarlosh2h.csv',usecols = [0,1,2], skiprows = 1, nrows = 19, delimiter =';')
mvcspts.columns = ['Country', 'MV', 'CS']
mvcspts
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))

#MV vs CS position w/ DNF, Line plot

wdnf = plt.subplot(2,2,1)
plt.plot(mvcspts['Country'], mvcspts['MV'], 'blue')
plt.plot(mvcspts['Country'], mvcspts['CS'], 'red')
plt.axhline(y = 0, linestyle = '-', color = 'orange')
plt.xticks(rotation = 90)
plt.title('MV vs CS position by race w dnf, 2015')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Carlos Sainz', 'DNF'])
plt.yticks(np.arange(0, 17, step = 1))
wdnf.invert_yaxis()

#MV vs CS position w/o DNF, Line plot

dnf = plt.subplot(2,2,2)
nodnf = mvcspts[(mvcspts.MV != 0) & (mvcspts.CS != 0)]
nodnf
plt.plot(nodnf['Country'], nodnf['MV'], 'blue')
plt.plot(nodnf['Country'], nodnf['CS'], 'red')
plt.xticks(rotation = 90)
plt.title('MV vs CS position by race w/o dnf, 2015')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Carlos Sainz'])
plt.yticks(np.arange(3, 17, step = 1))
dnf.invert_yaxis()

#MV vs CS position w/ DNF, scatter 

wdnf2 = plt.subplot(2,2,3)
plt.scatter(mvcspts['Country'], mvcspts['MV'], color = 'blue')
plt.scatter(mvcspts['Country'], mvcspts['CS'], color = 'red')
plt.axhline(y = 0, linestyle = '-', color = 'orange')
plt.xticks(rotation = 90)
plt.title('MV vs CS position by race w dnf, 2015')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Carlos Sainz', 'DNF'])
plt.yticks(np.arange(0, 17, step = 1))
wdnf2.invert_yaxis()

#MV vs CS position w/o DNF, scatter 

dnf2 = plt.subplot(2,2,4)
plt.scatter(nodnf['Country'], nodnf['MV'], color =  'blue')
plt.scatter(nodnf['Country'], nodnf['CS'], color = 'red')
plt.axhline(y = np.median(nodnf['MV']) , linestyle = 'dotted', color = 'green',linewidth = 2.5)
plt.axhline(y = np.median(nodnf['CS']) , linestyle = 'dotted', color = 'black', linewidth = 2.5)
plt.xticks(rotation = 90)
plt.title('MV vs CS position by race w/o dnf, 2015')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Carlos Sainz', 'MV Median Position', 'CS Median Position'])
plt.yticks(np.arange(3, 17, step = 1))
dnf2.invert_yaxis()


plt.tight_layout()
```

```python
#Max Verstappen vs Sergio Perez data - 2023
mvchecpos = pd.read_csv('mxchecoh2h.csv', delimiter =';', usecols = [0,1,2], nrows = 20)
mvchecpos.columns = ['Country', 'MV', 'SP']

#Data Exclusing DNFs 
nodnf2 = mvchecpos[(mvchecpos.MV != 0) & (mvchecpos.SP != 0)]
```

```python
plt.figure(figsize=(10, 8))

#MV vs SP w/ DNF, Line plot

wdnf3 = plt.subplot(2,2,1)
plt.plot(mvchecpos['Country'], mvchecpos['MV'], 'blue')
plt.plot(mvchecpos['Country'], mvchecpos['SP'], 'green')
plt.axhline(y = 0, linestyle = '-', color = 'orange')
plt.xticks(rotation = 90)
plt.title('MV vs SP position by race w dnf, 2023')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Sergio Perez', 'DNF'])
plt.yticks(np.arange(0, 17, step = 1))
wdnf3.invert_yaxis()

#MV vs SP w/o DNF, Line plot

nodnfs = plt.subplot(2,2,2)
plt.plot(nodnf2['Country'], nodnf2['MV'], 'blue')
plt.plot(nodnf2['Country'], nodnf2['SP'], 'green')
plt.axhline(y = np.median(nodnf2['MV']) , linestyle = 'dotted', color = 'red',linewidth = 2.5)
plt.axhline(y = np.median(nodnf2['SP']) , linestyle = 'dotted', color = 'orange', linewidth = 2.5)
plt.xticks(rotation = 90)
plt.title('MV vs SP position by race w/o dnf')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Sergio Perez', 'MV Median Position', 'SP Median Position'])
plt.yticks(np.arange(1, 17, step = 1))
nodnfs.invert_yaxis()

#MV vs SP w/ DNF, Scatter 

wdnf4 = plt.subplot(2,2,3)
plt.scatter(mvchecpos['Country'], mvchecpos['MV'], color = 'blue')
plt.scatter(mvchecpos['Country'], mvchecpos['SP'], color = 'green')
plt.axhline(y = 0, linestyle = '-', color = 'orange')
plt.xticks(rotation = 90)
plt.title('MV vs SP position by race w dnf, 2023')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Sergio Perez', 'DNF'])
plt.yticks(np.arange(0, 17, step = 1))
wdnf4.invert_yaxis()

#MV vs SP w/o DNF, Scatter

nodnfs2 = plt.subplot(2,2,4)
plt.scatter(nodnf2['Country'], nodnf2['MV'], color = 'blue')
plt.scatter(nodnf2['Country'], nodnf2['SP'], color = 'green')
plt.axhline(y = np.median(nodnf2['MV']) , linestyle = 'dotted', color = 'red',linewidth = 2.5)
plt.axhline(y = np.median(nodnf2['SP']) , linestyle = 'dotted', color = 'orange', linewidth = 2.5)
plt.xticks(rotation = 90)
plt.title('MV vs SP position by race w/o dnf, 2023')
plt.ylabel('position')
plt.grid()
plt.legend(['Max Verstappen', 'Sergio Perez', 'MV Median Position', 'SP Median Position'])
plt.yticks(np.arange(1, 17, step = 1))
nodnfs2.invert_yaxis()



plt.tight_layout()
```

```python
#MV vs SP boxplot w/o DNFS 

np.random.seed(10)
data = [nodnf2['MV'], nodnf2['SP']]
 
fig = plt.figure(figsize =(4, 4))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data, patch_artist = True, vert = 0)

for median in bp['medians']:
    median.set(color ='red',
               linewidth = 2)
    
colors = ['orange', 'pink']
 
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    
ax.set_yticklabels(['Max Verstappen', 'Sergio Perez'])
plt.title("MV vs SP Position w/o DNF, 2023")

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.xticks(np.arange(1, 18, step = 1))

plt.show()
```
